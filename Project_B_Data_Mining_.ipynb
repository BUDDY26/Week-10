```python
#
import numpy as np

data = np.load('coffee_combined_data.npz')
coffee_dataset = data['dataset']
coffee_labels = data['labels']

print(coffee_dataset)
print(coffee_labels)
```

```python
# Check the shape of the dataset and labels
print("Dataset shape:", coffee_dataset.shape)
print("Labels shape:", len(coffee_labels))  # Use len for label arrays
```

```python
# Check for NaN values in the dataset
print("Any NaN in dataset?", np.isnan(coffee_dataset).any())

# Check for empty or missing labels
missing_labels = any(label.strip() == '' for label in coffee_labels)
print("Any missing labels?", missing_labels)
```

```python
# Check unique labels and their counts
unique_labels = np.unique(coffee_labels)
print("Unique Labels:", unique_labels)
```

```python
import matplotlib.pyplot as plt

# Visualize the first sample in the dataset
time = np.arange(coffee_dataset.shape[1])  # Assuming second axis is time
plt.plot(time, coffee_dataset[0], label=f"Label: {coffee_labels[0]}")
plt.xlabel('Time')
plt.ylabel('Water Weight')
plt.title('Brewing Data - Single Sample')
plt.legend()
plt.show()

```

```python
# Plot all samples, color-coded by labels
for i in range(coffee_dataset.shape[0]):
    if coffee_labels[i] == 'Sprout':
        plt.plot(time, coffee_dataset[i], color='green', alpha=0.5, label='Sprout' if i == 0 else "")
    elif coffee_labels[i] == 'Rwanda':
        plt.plot(time, coffee_dataset[i], color='blue', alpha=0.5, label='Rwanda' if i == 0 else "")

plt.xlabel('Time')
plt.ylabel('Water Weight')
plt.title('Brewing Data - All Samples')
plt.legend()
plt.show()
```
